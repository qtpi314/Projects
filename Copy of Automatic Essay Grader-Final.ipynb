From the example above, we can see what the output is for a simple phrase that has been parsed and had the stopwords, punctuation, and anything that is not an alphanumeric (like @) removed. Now we are ready to apply it to all the essays in the scored essays data frame.

```
#function to parse the essays and extract the lemmas
def give_me_lemmas(text):
    phrase = nlp(text)
    return " ".join([x.lemma_ for x in phrase if not x.is_stop and not x.is_punct and x.is_alpha and x.pos_ != "SCONJ"])
```

Below is how I built the data frame using TensorFlow. I used the Universal Sentence Encoder (USE) embedded feature from the TensorFlow package. The Universal Sentence Encoder calculates 512 vectors that also represent each essay just like the SpaCy vector representation above. However, these vector representations take context into account, meaning the vectors are based on the sentences in each essay, not just the words. For example, if you had two sentences, each with the same words but in different order, the SpaCy vector representation would calculate them to be very similar. The USE representation would result in two different vectors.

Again, the data frame of the USE vectors was saved and uploaded to my GitHub repository. So that line of code has been commented out and no longer needed. However, I left it here for completion.

```
import tensorflow_hub as hub

embed = hub.load("https://tfhub.dev/google/universal-sentence-encoder/4")

def give_me_use_vectors(phrase):
    phrase_as_tf_tensor = embed([phrase])
    phrase_as_numpy_array = phrase_as_tf_tensor.numpy()
    phrase_only_what_I_need = phrase_as_numpy_array[0]
    return pd.Series(phrase_only_what_I_need)
```

```
df_of_use_vectors = scored_essays['essay'].apply(give_me_use_vectors)
```


# Part 2: Models using TF-IDF Representation
This section contains the method for performing TF-IDF vectorization with 2-word ngrams on the final preprocessed essays. TF-IDF stands for Text Frequency Inverse Document Frequency. I chose to use TF-IDF because it not only takes into account the occurances of the words, but also pays attention to semantics using ngrams. In general, the TF-IDF algorithm counts the frequency that a word appears in a document. If that word also appears in many other documents, that word is not essential to the overall meaning of that particular text. Words with a high frequency but are more unique to a document recieve a higher score because the word impacts the meaning of the overall text more so than frequent words that are common to many different texts. 

A random forest classifier as well as a regressor model was applied using the TF-IDF vectorization. Grid Search was used for hyper parameter tuning. The accuracy of the model is 86% which verifies the test score of the model. 

The precision score, which indicates the accuracy of a positive identification is 82%. So 82% of the observations identified as true are actually true.

The recall score, which indicates the percentage of positive observations that the model correctly identified is 65%. In other words, 65% of the truely positive observations were correctly identified. 

Ideally, we want both scores to be as high as possible. However this is rarely possible so we have to consider a tradeoff between the two. In this situation, I believe the recall score to be more important because if the model only passes 65% of students who ***should*** pass, in a production situation, you would have 35% of the good writers upset that their essay was failed.

The F1 score, which is the "Harmonic mean between precision and recall" scored a 72%. This is not bad as we want this value to be as close to 100% as possible.

The f-beta scores give weight to either the precision or the recall scores. Generally whichever score is higher recieves a higher f-beta score. In this case it is the f-beta favoring precision recieved a higher score because the precision score was higher than the recall score.

Lastly, we have the ROC curve. Ideally we want the "elbow" to be as close to the top left corner as possible. We want to minimize false positives while maximizing true positives. So we want the curve to be very steep at the beginning (small false positive rate) and level out for most of the graph. This curve indicates that the classification of the model is not merely random and the corner is sitting around the 85% accuracy which validates the confusion matrix and the test result of the model. This curve indicates the our model is fairly decent at identifying good writing from bad but it still can't provide an actual grade to an essay.

# Results and Recommendations
The results of all the models show that there is no significant difference between the different vector representations of the student essays. While the binary model did yield some relatively decent results in separating essays into pass and fail, this is nowhere near the point where it can be used to assist teachers with grading essays, which would save time and avoid teacher burn out. 

Much more research is required to improve the models to the point where they will be useful in a classroom setting. This can include collecting and analyzing more data and sets of essays. Possibly a wider range of essays from the lower grades through 12 grade and college would definitely represent more differences in writing ability. If we want an essay grader that can be used at all levels, we should include essays from all levels. We also want to allow the user to optimize it for their level and purposes so we should include some parameters and attributes that the user can tune themselves to make it appropriate to grade the right level essay. We don't want a 1st grade essay graded on the same standards as a college essay for example. We can only do this with a wider range of essays, thus demonstrating a wider range of abilities. Seeing a bigger variance in the data would possibly improve the overall performance of all the models. 

I would suggest using the Random Forest Classifier, simply for speed as there was no significant difference between that and the Regressor. As for the vector representations, again, no significant difference was found as all the models produced nearly identical results. I would probably choose either the SpaCy vetor represenation or the Universal Sentence Encoder because it uses less memory. The Universal Sentence Encoder is new and state of the art. It will most likely undergo the most significant changes and improvements moving forward so that would most likely be the better choice.
